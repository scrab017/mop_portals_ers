```python
import tabula
import pandas as pd
import glob
```

### Report 1

```python
for file in glob.glob('../data/raw/dailyPubReports/Report1/*.pdf'):
    
    temp_1 = tabula.read_pdf(file,pandas_options={'header':3})
    
    temp_1_n = temp_1.iloc[:,[0,3,4]][1:5]
    temp_1_n.columns = ['particulars','program_today', 'actual_today']
    temp_1_n['region'] = 'Northern'
    temp_1_n = temp_1_n[['region','particulars','program_today', 'actual_today']]
    
    temp_1_w = temp_1.iloc[:,[0,3,4]][6:10]
    temp_1_w.columns = ['particulars','program_today', 'actual_today']
    temp_1_w['region'] = 'Western'
    temp_1_w = temp_1_w[['region','particulars','program_today', 'actual_today']]
    
    temp_1_s = temp_1.iloc[:,[0,3,4]][11:15]
    temp_1_s.columns = ['particulars','program_today', 'actual_today']
    temp_1_s['region'] = 'Southern'
    temp_1_s = temp_1_s[['region','particulars','program_today', 'actual_today']]
    
    temp_1_e = temp_1.iloc[:,[0,3,4]][16:19]
    temp_1_e.columns = ['particulars','program_today', 'actual_today']
    temp_1_e['region'] = 'Eastern'
    temp_1_e = temp_1_e[['region','particulars','program_today', 'actual_today']]
    
    temp_1_ne = temp_1.iloc[:,[0,3,4]][20:23]
    temp_1_ne.columns = ['particulars','program_today', 'actual_today']
    temp_1_ne['region'] = 'North Eastern'
    temp_1_ne = temp_1_ne[['region','particulars','program_today', 'actual_today']]
    
    temp_1_ai = temp_1.iloc[:,[0,3,4]][24:]
    temp_1_ai.columns = ['particulars','program_today', 'actual_today']
    temp_1_ai['region'] = 'All India'
    temp_1_ai = temp_1_ai[['region','particulars','program_today', 'actual_today']]
    
    df_1 = pd.concat([temp_1_n,temp_1_w,temp_1_s,temp_1_e,temp_1_ne,temp_1_ai])
    
    df_1.program_today = df_1.program_today.apply(lambda x:float(x.replace(',','')) if type(x)!=float else float(x))
    df_1.actual_today = df_1.actual_today.apply(lambda x:float(x.replace(',','')) if type(x)!=float else float(x))
    
    df_1.to_csv('../data/processed/dailyPubReports/Report1/'+file.split('\\')[-1].split('.')[0]+'.csv',index=False)
```

### Report 3

```python
for file in glob.glob('../data/raw/dailyPubReports/Report3/*.pdf'):
    
    temp_3 = tabula.read_pdf(file,pandas_options={'header':3})[['1','2','3','4','5']]
    temp_3.columns = ['region','monitored_capacity_MW','planned_maintanence_MW','forced_MW','other_reasons_MW']
    
    temp_3.monitored_capacity_MW = temp_3.monitored_capacity_MW.apply(lambda x:float(x.replace(',','')))
    temp_3.planned_maintanence_MW = temp_3.planned_maintanence_MW.apply(lambda x:float(x.replace(',','')))
    temp_3.forced_MW = temp_3.forced_MW.apply(lambda x:float(x.replace(',','')))
    temp_3.other_reasons_MW = temp_3.other_reasons_MW.apply(lambda x:float(x.replace(',','')))
    
    temp_3.to_csv('../data/processed/dailyPubReports/Report3/'+file.split('\\')[-1].split('.')[0]+'.csv',index=False)
```

### Report 4

```python
for file in glob.glob('../data/raw/dailyPubReports/Report4/*.pdf'):
    
    temp_4_list = tabula.read_pdf(file,pages='all',multiple_tables=True)
    
    temp_4_0 = temp_4_list[0][[0,1,3,5,6,7,8]]

    temp_4_1 = temp_4_list[1][[0,1,3,5,6,7,8]]

    temp_4_2 = temp_4_list[2].drop(1,axis=1)[[0,2,4,6,7,8,9]]
    temp_4_2.columns = [0, 1, 3, 5, 6, 7, 8]

    temp_4_3 = temp_4_list[3][[0,1,3,5,6,7,8]]

    temp_4_4 = temp_4_list[4][[0,1,3,5,6,7,8]]

    temp_4_5 = temp_4_list[5].drop(1,axis=1)[[0,2,4,6,7,8,9]]
    temp_4_5.columns = [0, 1, 3, 5, 6, 7, 8]

    temp_4_n = pd.concat([temp_4_0[9:],temp_4_1[8:24]]).reset_index().drop('index',axis=1)
    temp_4_n['3_1'] = temp_4_n[3].apply(lambda x:x.split(' ')[0])
    temp_4_n['3_2'] = temp_4_n[3].apply(lambda x:x.split(' ')[1] if (' ') in x else 0.00)
    temp_4_n['7_1'] = temp_4_n[7].apply(lambda x:x.split(' ')[0])
    temp_4_n['7_2'] = temp_4_n[7].apply(lambda x:x.split(' ')[1] if (' ') in x else 0.00)
    temp_4_n[8] = temp_4_n[8].apply(lambda x:x.split(' ')[0])
    temp_4_n['region'] = 'Northern'
    temp_4_n = temp_4_n[['region',0,1,'3_1','3_2',5,6,'7_1','7_2',8]]
    temp_4_n.columns = ['region','power_station','monitored_capacity_MW','planned_maintanence_MW','forced_maintanence_MW',
                        'other_reasons_MW','gen_dt_program','gen_dt_actual','gen_dt_excess_shortfall','gen_dt_deviation']

    temp_4_w = pd.concat([temp_4_1[25:],temp_4_2[8:],temp_4_3[8:18]]).reset_index().drop('index',axis=1)
    temp_4_w['3_1'] = temp_4_w[3].apply(lambda x:x.split(' ')[0])
    temp_4_w['3_2'] = temp_4_w[3].apply(lambda x:x.split(' ')[1] if (' ') in x else 0.00)
    temp_4_w['7_1'] = temp_4_w[7].apply(lambda x:x.split(' ')[0])
    temp_4_w['7_2'] = temp_4_w[7].apply(lambda x:x.split(' ')[1] if (' ') in x else 0.00)
    temp_4_w[8] = temp_4_w[8].apply(lambda x:x.split(' ')[0])
    temp_4_w['region'] = 'Western'
    temp_4_w = temp_4_w[['region',0,1,'3_1','3_2',5,6,'7_1','7_2',8]]
    temp_4_w.columns = ['region','power_station','monitored_capacity_MW','planned_maintanence_MW','forced_maintanence_MW',
                        'other_reasons_MW','gen_dt_program','gen_dt_actual','gen_dt_excess_shortfall','gen_dt_deviation']


    temp_4_s = pd.concat([temp_4_3[19:],temp_4_4[8:26]]).reset_index().drop('index',axis=1)
    temp_4_s['3_1'] = temp_4_s[3].apply(lambda x:x.split(' ')[0])
    temp_4_s['3_2'] = temp_4_s[3].apply(lambda x:x.split(' ')[1] if (' ') in x else 0.00)
    temp_4_s['7_1'] = temp_4_s[7].apply(lambda x:x.split(' ')[0])
    temp_4_s['7_2'] = temp_4_s[7].apply(lambda x:x.split(' ')[1] if (' ') in x else 0.00)
    temp_4_s[8] = temp_4_s[8].apply(lambda x:x.split(' ')[0])
    temp_4_s['region'] = 'Southern'
    temp_4_s = temp_4_s[['region',0,1,'3_1','3_2',5,6,'7_1','7_2',8]]
    temp_4_s.columns = ['region','power_station','monitored_capacity_MW','planned_maintanence_MW','forced_maintanence_MW',
                        'other_reasons_MW','gen_dt_program','gen_dt_actual','gen_dt_excess_shortfall','gen_dt_deviation']


    temp_4_e = pd.concat([temp_4_4[27:],temp_4_5[8:25]]).reset_index().drop('index',axis=1)
    temp_4_e[0] = temp_4_e[0].fillna('Total')
    temp_4_e['3_1'] = temp_4_e[3].apply(lambda x:x.split(' ')[0])
    temp_4_e['3_2'] = temp_4_e[3].apply(lambda x:x.split(' ')[1] if (' ') in x else 0.00)
    temp_4_e['7_1'] = temp_4_e[7].apply(lambda x:x.split(' ')[0])
    temp_4_e['7_2'] = temp_4_e[7].apply(lambda x:x.split(' ')[1] if (' ') in x else 0.00)
    temp_4_e[8] = temp_4_e[8].apply(lambda x:x.split(' ')[0])
    temp_4_e['region'] = 'Eastern'
    temp_4_e = temp_4_e[['region',0,1,'3_1','3_2',5,6,'7_1','7_2',8]]
    temp_4_e.columns = ['region','power_station','monitored_capacity_MW','planned_maintanence_MW','forced_maintanence_MW',
                        'other_reasons_MW','gen_dt_program','gen_dt_actual','gen_dt_excess_shortfall','gen_dt_deviation']


    temp_4_ne = temp_4_5[26:].reset_index().drop('index',axis=1)
    temp_4_ne[0] = temp_4_ne[0].fillna('Total')
    temp_4_ne['3_1'] = temp_4_ne[3].apply(lambda x:x.split(' ')[0])
    temp_4_ne['3_2'] = temp_4_ne[3].apply(lambda x:x.split(' ')[1] if (' ') in x else 0.00)
    temp_4_ne['7_1'] = temp_4_ne[7].apply(lambda x:x.split(' ')[0])
    temp_4_ne['7_2'] = temp_4_ne[7].apply(lambda x:x.split(' ')[1] if (' ') in x else 0.00)
    temp_4_ne[8] = temp_4_ne[8].apply(lambda x:x.split(' ')[0])
    temp_4_ne['region'] = 'North Eastern'
    temp_4_ne = temp_4_ne[['region',0,1,'3_1','3_2',5,6,'7_1','7_2',8]]
    temp_4_ne.columns = ['region','power_station','monitored_capacity_MW','planned_maintanence_MW','forced_maintanence_MW',
                        'other_reasons_MW','gen_dt_program','gen_dt_actual','gen_dt_excess_shortfall','gen_dt_deviation']

    df_4 = pd.concat([temp_4_n,temp_4_w,temp_4_s,temp_4_e,temp_4_ne])
    
    df_4.monitored_capacity_MW = df_4.monitored_capacity_MW.apply(lambda x:float(x.replace(',','')) if type(x)!=float else float(x))
    df_4.planned_maintanence_MW = df_4.planned_maintanence_MW.apply(lambda x:float(x.replace(',','')) if type(x)!=float else float(x))
    df_4.forced_maintanence_MW = df_4.forced_maintanence_MW.apply(lambda x:float(x.replace(',','')) if type(x)!=float else float(x))
    df_4.other_reasons_MW = df_4.other_reasons_MW.apply(lambda x:float(x.replace(',','')) if type(x)!=float else float(x))
    df_4.gen_dt_program = df_4.gen_dt_program.apply(lambda x:float(x.replace(',','')) if type(x)!=float else float(x))
    df_4.gen_dt_actual = df_4.gen_dt_actual.apply(lambda x:float(x.replace(',','')) if type(x)!=float else float(x))
    df_4.gen_dt_excess_shortfall = df_4.gen_dt_excess_shortfall.apply(lambda x:float(x.replace(',','')) if type(x)!=float else float(x))
    df_4.gen_dt_deviation = df_4.gen_dt_deviation.apply(lambda x:float(x.replace(',','')) if type(x)!=float else float(x))
    
    df_4.to_csv('../data/processed/dailyPubReports/Report4/'+file.split('\\')[-1].split('.')[0]+'.csv',index=False)
```

### Report 17

```python
for file in glob.glob('../data/raw/dailyPubReports/Report17/*.pdf'):
    
    temp_17 = tabula.read_pdf(file,pandas_options={'header':3})
    
    temp_17_thermal = temp_17[1:8][['1','2','3','4']]
    temp_17_nuclear = temp_17[9:10][['1','2','3','4']]
    temp_17_hydro = temp_17[11:13][['1','2','3','4']]
    temp_17_total = temp_17[13:14][['1','2','3','4']]

    temp_17_thermal.columns = ['particulars_1','monitored_capacity_MW','program_MU','actual_MU']
    temp_17_thermal['particulars_0'] = 'THERMAL'
    temp_17_thermal = temp_17_thermal[['particulars_0','particulars_1','monitored_capacity_MW','program_MU','actual_MU']]

    temp_17_nuclear.columns = ['particulars_1','monitored_capacity_MW','program_MU','actual_MU']
    temp_17_nuclear['particulars_0'] = 'NUCLEAR'
    temp_17_nuclear = temp_17_nuclear[['particulars_0','particulars_1','monitored_capacity_MW','program_MU','actual_MU']]

    temp_17_hydro.columns = ['particulars_1','monitored_capacity_MW','program_MU','actual_MU']
    temp_17_hydro['particulars_0'] = 'HYDRO'
    temp_17_hydro = temp_17_hydro[['particulars_0','particulars_1','monitored_capacity_MW','program_MU','actual_MU']]

    temp_17_total.columns = ['particulars_1','monitored_capacity_MW','program_MU','actual_MU']
    temp_17_total['particulars_0'] = 'TOTAL'
    temp_17_total = temp_17_total[['particulars_0','particulars_1','monitored_capacity_MW','program_MU','actual_MU']]

    df_17 = pd.concat([temp_17_thermal,temp_17_nuclear,temp_17_hydro,temp_17_total])

    df_17.monitored_capacity_MW = df_17.monitored_capacity_MW.apply(lambda x:float(x.replace(',','')))
    df_17.program_MU = df_17.program_MU.apply(lambda x:float(x.replace(',','')))
    df_17.actual_MU = df_17.actual_MU.apply(lambda x:float(x.replace(',','')))
    
    df_17.to_csv('../data/processed/dailyPubReports/Report17/'+file.split('\\')[-1].split('.')[0]+'.csv',index=False)
```
